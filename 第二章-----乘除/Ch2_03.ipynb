## 标量相乘

```python
import numpy as np
row_a = np.array([[1,2,3]]) #创建一个行向量
b_col = np.array([[1],[2],[3]]) #创建一个列向量
A = np.array([[1, 2, 3],
              [4, 5, 6]]) #创建一个矩阵
```

```python
print(f"原向量为{row_a}乘一个标量2后结果为：{row_a*2}")
print(f"原向量为\n{b_col}\n乘一个标量2后结果为：\n{b_col*2}")
print(f"原矩阵为\n{A}\n乘一个标量2后结果为：\n{A*2}")
```

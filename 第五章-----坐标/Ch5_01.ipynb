```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4,0,-5,-4,6,0]
y = [2,-2,7,-6,-5,0]
labels = ['A','B','C','D','E','O']
fig ,ax = plt.subplots()
plt.plot(x,y,'x')
for lable ,i,j in zip(labels,x,y):
    plt.text(i,j+0.5,lable+ ' ({}, {})'.format(i, j))
    '''
    在代码 plt.text(i, j+0.5, label + ' ({}, {})'.format(i, j)) 中，j+0.5 是为了让文本标签显示在数据点的上方一定距离的位置。

默认情况下，plt.text()函数的第二个参数指定要显示文本的垂直坐标位置，该位置是以数据坐标系为准的。如果将文本标签与数据点末端对齐，

则可能会导致文本标签重叠或覆盖在数据点上，影响图形的美观度和可读性。因此，将文本标签垂直偏移一定距离可以更好地区分数据点和文本标签。

在这个例子中，由于每一项偏移量都是固定的0.5单位，因此数值大小并不影响结果，只要能够让标签稍稍从数据点“脱离”出来即可。

当然，具体的偏移量需要根据实际需求进行调整，以使得文本标签看起来既不过于拥挤也不过于分散。
    '''
plt.xlabel('x');plt.ylabel('y')
plt.axhline(y = 0,color = 'k',linestyle = '-')     #y轴
plt.axvline(x = 0,color = 'k',linestyle = '-')     #x轴
plt.xticks(np.arange(-8,8 + 1,step = 1))           #设置x轴刻度及标签
plt.yticks(np.arange(-8,8 + 1,step = 1))           #设置y轴刻度及标签
plt.axis('scaled')      #设置放缩比例相等

ax.set_xlim(-8,8);ax.set_ylim(-8,8)               #设置x,y轴范围
ax.spines['top'].set_visible(False);ax.spines['bottom'].set_visible(False)    #设置边框属性
ax.spines['left'].set_visible(False);ax.spines['right'].set_visible(False)

ax.grid(linestyle='--', linewidth=0.25, color=[0.5,0.5,0.5])    #添加啊网格线
```

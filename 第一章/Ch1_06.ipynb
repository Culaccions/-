```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

A_first_col = A[:,0] # saved as one dimension row
#A[:,0]是numpy中数组的一种写法，表示对一个二维数组，取该二维数组第一维中的所有数据，第二维中取第0个数据，直观来说，X[:,0]就是取所有行的第0个数据

A_first_col_V2 = A[:,[0]] # saved as a column

A_first_second_col_V2 = A[:,[0,1]] # extract first and second columns(一二列)

A_first_third_col_V2 = A[:,[0,2]] # extract first and third columns(一三列)

A_first_row = A[[0],:] # extract first row(一行)

A_second_row = A[[1],:] # extract second row(二行)

A_second_row_first_col = A[[1],[0]] # i = 2, j = 1
```

```python
print(A_first_col)
print("\n")
print(A_first_col_V2)
print("\n")
print(A_first_second_col_V2)
print("\n")
print(A_first_third_col_V2)
print("\n")
print(A_first_row)
print("\n")
print(A_second_row)
print("\n")
print(A_second_row_first_col)
```

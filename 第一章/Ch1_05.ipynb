```python
import numpy as np
```

```python
a_row = np.array([[1,2,3]])        #如果写成np.array([1,2,3])则表示的是一个维度，不能表示一个向量，几个大括号就是几维
```

```python
print(a_row)
```

```python
b = a_row.T
```

对a_row进行转置操作

```python
print(b)
```

```python
b_col = np.array([[1],[2],[3]])
```

```python
print(b_col)
```

```python
a = b_col.T
```

对b_col进行转置操作

```python
print(a)
```

.T表示转置操作，行向量.T得到的为列向量，列向量.T转置成了行向量
